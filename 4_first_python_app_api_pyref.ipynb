```python
# Converting years to months
```

```python
user_age = input("Enter your age (in years): ")
years = int(user_age)
months = years * 12
print(f"Your age, {years}, is equal to {months} months")
```

```python
# Converting years to seconds
```

```python
user_age = input("Enter your age (in years): ")
years = int(user_age)
months = years * 12
days = years * 365  # approximating 365 days for a year
hours = days * 24
minutes = hours * 60
seconds = minutes * 60
print(f"Your age, {years}, is equal to {seconds} seconds")
```
