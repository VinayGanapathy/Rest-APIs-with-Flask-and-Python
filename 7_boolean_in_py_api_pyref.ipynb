```python
# Applicable On strings, lists, tuples etc
```

```python
friends = ["Rolf", "Bob"]
abroad = ["Rolf", "Bob"]
```

```python
print(friends == abroad)
```

```python
print(friends is abroad) # Eventhough the two lists have the same elements they occupy different memory allocations
```

```python
friends = ["Rolf", "Bob"]
abroad = friends
```

```python
print(friends is abroad) # Here the two lists have the same elements and they occupy same memory allocation.
```
