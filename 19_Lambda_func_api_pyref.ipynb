```python
# Lambda Function in Python
```

```python
def add(x,y):
  return x + y
```

```python
print(add(5,7))
```

```python
my_add = lambda x,y : x + y
```

```python
print(my_add(5,7))
```

```python
def double(x):
  return x * 2
```

```python
sequence = [1,3,5,7]
doubled = [x * 2 for x in sequence]
print(doubled)
```

```python
sequence = [1,3,5,7]
doubled = [double(x) for x in sequence]
print(doubled)
```

```python
sequence = [1,3,5,7]
doubled = map (double, sequence)
print(list(doubled))
```

```python
sequence = [1,3,5,7]
doubled = [(lambda x : x * 2)(x) for x in sequence]
print(doubled)
```

```python
sequence = [1,3,5,7]
doubled = list(map(lambda x : x * 2, sequence))
print(doubled)
```
