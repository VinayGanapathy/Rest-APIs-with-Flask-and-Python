```python
# Default function parameter value
```

```python
def add(x,y):
  print(x + y)
```

```python
add(5, 8)
```

```python
def add(x,y = 8):
  print(x + y)
```

```python
add(5, 10)         # 'x' is the required parameter and 'y' is the optional parameter
```

```python
add(5) 
```

```python
add(x = 5) 
```

```python
add(y = 5) 
```

```python
# default parameters must go at the end
```

```python
default_y = 3
def add(x, y = default_y):
  sum = x + y
  print(sum)
```

```python
add(2)
```

```python
default_y = 4
add(2)             # This doesn't change the default_y value inside the function    
```

```python
def add(x, y = 3):
  sum = x + y
  print(sum)                   # This is the correct way of defining the default values to the function
```
