```python
def add(x,y):
  print(x+y)

print(add(5,8))
```

```python
def add(x,y):
  print(x+y)
  return x + y               # return should always be put at the end of the function.
```

```python
result = add(5,8)
print(result)
```

```python
def divide(dividend, divisor):
  if divisor != 0 :
    return dividend/divisor
  else:
    return "Not possible !"
```

```python
result = divide(15,3)
print(result)
```

```python
result = divide(15,0)
print(result)
```
