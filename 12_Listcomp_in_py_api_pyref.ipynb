```python
# List comprehensions
```

```python
numbers = [1,3,5]
```

```python
doubled = []
for num in numbers:
  doubled.append(num*2)
print(doubled)
```

```python
# Using list comprehension
doubled_l = [num * 2 for num in numbers]
print(doubled_l)
```

```python
# another example
friends = ["Rolf", "Sam", "Samantha", "Saurabh", "Jen"]
starts_s = []

for friend in friends:
  if friend.startswith("S"):
    starts_s.append(friend)

print(starts_s)
```

```python
starts_l=[friend for friend in friends if friend.startswith("S")]
print(starts_l)
```

```python
print(starts_s is starts_l)
```

```python
print("friends", id(friends))   # id is the memorry allocation address
```
