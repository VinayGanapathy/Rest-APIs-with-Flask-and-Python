```python
# Unpacking keyword arguments
```

```python
def named(**kwargs):
  print(kwargs)
```

```python
named(name = "Bob", age = "25")
```

```python
def named(name, age):
  print(name, age)

details = {"name":"Bob", "age":25 }
named(details)
```

```python
def named(name, age):
  print(name, age)

details = {"name":"Bob", "age":25 }
named(**details)
```

```python
def named(**kwargs):
  print(kwargs)

details = {"name":"Bob", "age":25 }
named(**details)
```

```python
def named(**kwargs):
  print(kwargs)

def print_nicely(**kwargs):
  named(**kwargs)
  for arg,value in kwargs.items():
    print(f"{arg}:{value}")

details = {"name":"Bob", "age":25 }
named(**details)
```

```python
print_nicely(**details)
```

```python
def both(*args, **kwargs):
  print(args)
  print(kwargs)
```

```python
both(1,3,5, name="Bob", age = 25)
```
