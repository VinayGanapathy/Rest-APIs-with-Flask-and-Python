```python
# String formatting in python
```

```python
name = 'Bob'

greeting = 'Hello, Bob'

print(greeting)
```

```python
name = 'Rolf'
print(greeting)
```

```python
greeting = f"Hello, {name}"
```

```python
print(greeting)
```

```python
# Template using .format
```

```python
name = 'Bob'
greeting = 'Hello, {}'
with_name = greeting.format(name)
with_name_two = greeting.format('Rolf')
```

```python
print(with_name)
```

```python
print(with_name_two)
```

```python
longer_phrase = "Hello, {}. Today is {}"

formatted = longer_phrase.format("Rolf", "Monday")

print(formatted)
```
