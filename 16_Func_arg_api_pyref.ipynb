```python
# Function arguments and parameters
```

```python
def add(x,y):
  pass                              # pass means 'do nothing'
```

```python
add(5,3)                       #'5' and '3' are called parameters or arguments
```

```python
def add(x,y):
  result = x + y
  print(result)
```

```python
add(5,3)  # This is called as positional arguments
```

```python
def say_hello():
  print("Hello")
```

```python
say_hello("Bob")   # arguments are not allowed
```

```python
say_hello() 
```

```python
def say_hello(name):
  print(f"Hello, {name}")
```

```python
say_hello("Bob")
```

```python
say_hello()
```

```python
def say_hello(name, surname):
  print(f"Hello, {name} {surname}")
```

```python
say_hello("Bob", "Smith")                 # This is called as positional arguments
```

```python
say_hello(surname = "Smith", name = "Bob") # This is called as keyword arguments
```

```python
def divide (dividend, divisor):
  if divisor != 0:
    print(dividend/divisor)
  else:
    print("Not possible!")
```

```python
divide(15,0)
```

```python
divide (divisor = 0, dividend = 15)
```

```python
# Positional arguments always comes first before keyword arguments
# Using keyword arguments is always recommended.
```
